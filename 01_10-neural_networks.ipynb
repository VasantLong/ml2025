## 人工神经网络(Artificial neural network)

对人类中枢神经系统的观察启发了人工神经网络这个概念。在人工神经网络中，简单的人工节点，称作神经元（neurons），连接在一起形成一个类似生物神经网络的网状结构。一个神经元通常具有多个树突，主要用来接受传入信息；而轴突只有一条，轴突尾端有许多轴突末梢可以给其他多个神经元传递信息。轴突末梢跟其他神经元的树突产生连接，从而传递信号。这个连接的位置在生物学上叫做“突触”。

在神经网络的现代软件实现中，被生物学启发的那种方法已经很大程度上被抛弃了，取而代之的是基于统计学和信号处理的更加实用的方法。从历史的角度讲，神经网络模型的应用标志着二十世纪八十年代后期从**高度符号化的人工智能**（以用条件规则表达知识的专家系统为代表）向低符号化的机器学习（以用动力系统的参数表达知识为代表）的转变。

典型的人工神经网络具有以下三个部分：
* **神经元** ANNs是由人工神经元组成的，它保留了生物学上的神经元的概念，神经元接受输入，将输入与其内部状态（激活）和可选的阈值结合在一起，使用激活函数，使用输出函数产生输出。最初的输入是外部数据，如图像、文档等外部数据。最终输出完成任务，如识别图像中的物体等。激活函数的重要特点是，当输入值发生变化时，它提供了一个平滑的、可区分的过渡，即输入的微小变化会产生微小的输出变化。
> 在神经网络中，激活函数决定来自给定输入集的节点的输出，其中非线性激活函数允许网络复制复杂的非线性行为。如果不用激励函数，其实相当于激励函数是恒等函数$f(x) = x$，在这种情况下你每一层节点的输入都是上层输出的线性函数，很容易验证，无论你神经网络有多少层，输出都是输入的线性组合，与没有隐藏层效果相当，这种情况就是最原始的感知机（Perceptron）了，那么网络的逼近能力就相当有限。正因为上面的原因，引入非线性函数作为激励函数，这样深层神经网络表达能力就更加强大（不再是输入的线性组合，而是几乎可以逼近任意函数）。

* **连接和权重** 神经网络由连接组成，每个连接提供一个神经元的输出作为另一个神经元的输入。每个连接都被分配了一个权重，代表其相对重要性。一个给定的神经元可以有多个输入和输出连接。
* **传播函数** 传播函数从其前代神经元的输出及其连接计算出一个神经元的输入，作为加权和，也可以在传播结果中加入一个偏置项。

### 网络结构

神经元通常被组织成多个层，尤其是在深度学习(deep learning)中。一个层的神经元只与紧随其后的层的神经元连接。接收外部数据的层是输入层(input layer)。产生最终结果的层是输出层(output layer)。在它们之间是零层或更多的隐藏层。也有单层网络和非分层网络。在两层之间，可以有多种连接模式。它们可以是完全连接的( fully connected)，一层中的每个神经元与下一层中的每个神经元连接。它们的连接模式可以是池化(pool)，其中一层的一组神经元连接到下一层的*单个*神经元，从而减少该层的神经元的数量，这种连接的神经元形成一个有方向的无环形图（DAG），被称为前馈网络。 [45]另外，允许同一层或前一层的神经元之间连接的网络被称为循环神经网络（Recurrent neural networks）。

# 常用的激活函数(Activation function)

1. 非线性： 当激活函数是线性的时候，一个两层的神经网络就可以逼近基本上所有的函数了。但是，如果激活函数是恒等激活函数的时候（即f(x)=x），就不满足这个性质了，而且如果MLP使用的是恒等激活函数，那么其实整个网络跟单层神经网络是等价的。
1. 单调性： 当激活函数是单调的时候，单层网络能够保证是凸函数。
1. $f(x) \approx x$ 当激活函数满足这个性质的时候，如果参数的初始化是random的很小的值，那么神经网络的训练将会很高效；如果不满足这个性质，那么就需要很用心的去设置初始值。
1. 梯度消失问题：神经网络是利用梯度下降过程来训练的。梯度下降由基于链式规则的反向传播组成，链式规则用于获取权值变化以减少每次训练后的损失。链式规则在反向传播时会经过很多个层。换言之，因为网络的深度和激活过程将值转变为趋近于 0，它们的梯度逐渐消失，这就是所谓的梯度消失问题。所以我们希望理想的激活函数不要使梯度逐渐变为零。
1. 以零为中心：激活函数的输出应对称于零，这样梯度就不会向特定方向移动。
1. 可微性：当优化方法是基于梯度的时候，这个性质是必须的。
1. 输出值的范围： 当激活函数输出值是**有限**的时候，基于梯度的优化方法会更加**稳定**，因为特征的表示受有限权值的影响更显著；当激活函数的输出是 无限 的时候，模型的训练会更加高效，不过在这种情况小，一般需要更小的learning rate.
1. 计算成本：网络的每一层都会应用激活函数，它在深层网络中需要计算数百万次。因此，激活函数的计算成本应该很低。








#### Sigmoid/Logistic函数


$$
\begin{aligned}
\text{sigmoid}(x) = \sigma(x) &= \frac{1}{1+e^{-x}} \\
\sigma'(x) &= \sigma(x)(1 - \sigma(x)) = \frac{e^{-x}}{(e^{-x}+1)^2}
\end{aligned}
$$

Sigmoid函数现在很少在实际模型中使用，缺点是计算量大，会导致梯度消失问题且不以零为中心，通常在二进制分类问题中才会使用。


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, num=300)

fig, axes = plt.subplots(1, 2, figsize=(10,3), sharex=True)

s = 1/(1+np.exp(-x))
axes[0].plot(x, 1 / (1 + np.exp(-x)))
axes[1].plot(x, s * (1 - s))

axes[0].axhline(0, c="grey", ls=":", lw=1)
axes[0].axvline(0, c="grey", ls=":", lw=1)
axes[1].axhline(0, c="grey", ls=":", lw=1)
axes[1].axvline(0, c="grey", ls=":", lw=1)
fig.tight_layout()
```

#### tanh激活函数

$$
\begin{aligned}
\tanh(x) &= \frac{2}{1+e^{-2x}} - 1 = \frac {e^{x} - e^{-x}}{e^{x} + e^{-x}} \\
\tanh'(x) &= 1 - \tanh^2(x)
\end{aligned}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, num=300)

fig, axes = plt.subplots(1, 2, figsize=(10,3), sharex=True)

s = 2/(1+np.exp(-2*x)) - 1
axes[0].plot(x, s)
axes[1].plot(x, 1 - s**2)

axes[0].axhline(0, c="grey", ls=":", lw=1)
axes[0].axvline(0, c="grey", ls=":", lw=1)
axes[1].axhline(0, c="grey", ls=":", lw=1)
axes[1].axvline(0, c="grey", ls=":", lw=1)
fig.tight_layout()
```

#### RELU

修正线性单元（Rectified linear unit，ReLU）是神经网络中最常用的激活函数。它保留了 step 函数的生物学启发（只有输入超出阈值时神经元才激活），不过当输入为正的时候，导数不为零，从而允许基于梯度的学习（尽管在 x=0 的时候，导数是未定义的）。使用这个函数能使计算变得很快，因为无论是函数还是其导数都不包含复杂的数学运算。然而，当输入为负值的时候，ReLU 的学习速度可能会变得很慢，甚至使神经元直接无效，因为此时输入小于零而梯度为零，从而其权重无法得到更新，在剩下的训练过程中会一直保持静默。

$$
f(x) ={\begin{cases}0 & {\text{for }}x \leq 0 \\ x & {\text{for }}x>0\end{cases}}
$$
$$
f'(x)={\begin{cases}0&{\text{for }}x\leq 0\\1&{\text{for }}x>0\end{cases}}
$$

* 优点
    * 不会出现梯度消失，收敛速度快；
    * 前向计算量小，只需要计算max(0, x)，不像sigmoid中有指数计算；
    * 反向传播计算快，导数计算简单，无需指数、出发计算；
    * 有些神经元的值为0，使网络具有saprse性质，可减小过拟合。

* 缺点
    比较脆弱，在训练时容易“die”，反向传播中如果一个参数为0，后面的参数就会不更新。需要谨慎设置学习速率，不能设置太高。


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-6, 6, num=300)

fig, axes = plt.subplots(1, 2, figsize=(10,3), sharex=True)

relu = np.vectorize(lambda x: x if x > 0 else 0, otypes=["float64"])
d_relu = np.vectorize(lambda x: 1 if x > 0 else 0, otypes=["float64"])

axes[0].plot(x, relu(x), lw=2.5)
axes[1].plot(x, d_relu(x), lw=2.5)

axes[0].axhline(0, c="grey", ls=":", lw=1)
axes[0].axvline(0, c="grey", ls=":", lw=1)
axes[1].axhline(0, c="grey", ls=":", lw=1)
axes[1].axvline(0, c="grey", ls=":", lw=1)
fig.tight_layout()
```

#### 高斯误差线性单元, Gaussian Error Linear Units (GELU)

高斯误差线性单元激活函数在最近的 Transformer 模型（谷歌的 BERT 和 OpenAI 的 GPT-2）中得到了应用，能避免梯度消失问题。GELU 的论文来自 2016 年，但直到最近才引起关注。

$$
\begin{aligned}
\text{GELU}(x)  &= x P(X \le x) \\
\text{GELU}'(x) &= P(X \le x) + x P'(X \le x) \\
\end{aligned}
$$

上式中$P(X \le x)$是正态分布$\mathcal{N}\left(0, 1\right)$ 的累积分布函数(CDF)，$P'(X \le x)$是密度函数(PDF)。

```python
from scipy.stats import norm

def gelu(x):
    return x * norm.cdf(x)

def deriv_gelu(x):
    return norm.cdf(x) + x * norm.pdf(x)

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, num=300)

fig, axes = plt.subplots(1, 2, figsize=(10,3), sharex=True)

relu = np.vectorize(lambda x: x if x > 0 else 0, otypes=["float64"])
d_relu = np.vectorize(lambda x: 1 if x > 0 else 0, otypes=["float64"])

axes[0].plot(x, gelu(x), lw=2.5)
axes[1].plot(x, deriv_gelu(x), lw=2.5)

axes[0].axhline(0, c="grey", ls=":", lw=1)
axes[0].axvline(0, c="grey", ls=":", lw=1)
axes[1].axhline(0, c="grey", ls=":", lw=1)
axes[1].axvline(0, c="grey", ls=":", lw=1)
axes[1].axhline(0.5, c="grey", ls=":", lw=1)
axes[1].axhline(1, c="grey", ls=":", lw=1)

fig.tight_layout()
```

```python
def plot_nn_mlp(layers, node_size=0.1, fontsize=16, input_labels=None, output_labels=None, edge_labels=None, ax=None):
    
    try:
        import viznet
    except:
        print("ERROR: 需要安装viznet包")
        return
    
    if ax is None:
        ax = plt.gca()
    
    ax.axis("off")
    ax.axis("equal")
    
    max_n = max(*layers)
    layer_nodes = []
    nodes = []
    brush_input = viznet.NodeBrush('nn.input', size=0.1, ax=ax)
    brush_output = viznet.NodeBrush('nn.output', size=0.1, ax=ax)
    brush_hidden = viznet.NodeBrush('nn.hidden', size=0.1, ax=ax)
    edge = viznet.EdgeBrush('--', lw=2, ax=ax)

    def plot_layer(brush, layer_idx, n_node):
        nodes = []
        for i in range(n_node):
            node = brush >> (layer_idx, 1 + (i + 1)*(max_n - 1)/(n_node + 1))
            nodes.append(node)

        return nodes
    
    layer_input_idx = 0
    nodes = plot_layer(brush_input, layer_input_idx, layers[0])
    layer_nodes.append(nodes)
    
    layer_hidden_idx = 1
    for n_node in layers[1:-1]:
        nodes = plot_layer(brush_hidden, layer_hidden_idx, n_node)
        layer_hidden_idx += 1
        layer_nodes.append(nodes)

    layer_output_idx = len(layers) - 1
    nodes = plot_layer(brush_output, layer_output_idx, layers[layer_output_idx])
    layer_nodes.append(nodes)
    
    for layer_idx in range(1, len(layer_nodes)):
        if edge_labels and layer_idx <= len(edge_labels):
            _edge_labels = edge_labels[layer_idx - 1]
        else:
            _edge_labels = None
        
        src_nodes = layer_nodes[layer_idx - 1]
        n_src_nodes = len(src_nodes)
        for src_idx in range(n_src_nodes):
            node_src = src_nodes[n_src_nodes - 1 - src_idx]
            
            dst_nodes = layer_nodes[layer_idx]
            n_dst_nodes = len(dst_nodes)
            for dst_idx in range(n_dst_nodes):
                node_dst = dst_nodes[n_dst_nodes - 1 - dst_idx]
                el = edge >> (node_src, node_dst)
                if _edge_labels:
                    _label = _edge_labels[src_idx][dst_idx]
                    if isinstance(_label, str):
                        el.text(_label)
                    elif isinstance(_label, tuple):
                        _text, _pos = _label
                        el.text(_text, position=_pos)
                
    if input_labels is not None:
        nodes = layer_nodes[layer_input_idx]
        n_nodes = len(nodes)
        for idx, label in enumerate(input_labels):
            nodes[n_nodes - 1 - idx].text(label, 'center', fontsize=fontsize)

    if output_labels is not None:
        nodes = layer_nodes[layer_output_idx]
        n_nodes = len(nodes)
        for idx, label in enumerate(output_labels):
            nodes[n_nodes - 1 - idx].text(label, 'center', fontsize=fontsize)
            
```

```python
from sklearn.metrics import accuracy_score

def mpl_model(X, y, X_test, y_test, hidden_layer_sizes, activation, ax=None):
    model = MLPClassifier(hidden_layer_sizes=hidden_layer_sizes, 
                          activation=activation, 
                          solver='lbfgs', # 数据量较小时，lbfgs最优化算法更适合
                          max_iter=5000,
                          random_state=0)
    
    model.fit(X, y)
    from plot_helper import plot_2d_separator

    plot_2d_separator(model, X, fill=True, alpha=.3, ax=ax)
    ax.scatter(X[y==0, 0], [X[y==0, 1]], label="0", c="red",alpha=0.3,  marker="o")
    ax.scatter(X_test[y_test==0, 0], [X_test[y_test==0, 1]], c="orange", alpha=0.3, marker="o", label="test_0")
    ax.scatter(X[y==1, 0], [X[y==1, 1]], label="1", alpha=0.3, c="blue", marker="^")
    ax.scatter(X_test[y_test==1, 0], [X_test[y_test==1, 1]], c="green", marker="^", alpha=0.3,  label="test_1")
    ax.set_xlabel(f"hidden layer: {repr(hidden_layer_sizes)}    "
                  f"activation: {activation}   "
                  f"accuracy: {accuracy_score(y, model.predict(X)) : 0.2f}"
                  f", {accuracy_score(y_test, model.predict(X_test)) : 0.2f}")
    ax.legend()
    
```

**正则化**是一种防止过度学习的技术，在误差函数中加入约束权重。

$$
L'(\omega)=\frac{\alpha}{2}\| \omega \|_2^2 + L(\omega)
$$


`alpha`的设置

```python
X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

fig, axes = plt.subplots(2, 4, figsize=(20, 8))
for axx, n_hidden_nodes in zip(axes, [10, 100]):
    for ax, alpha in zip(axx, [0, 0.0001, 0.01,  1]):
        mlp = MLPClassifier(solver='lbfgs', random_state=0,
                            hidden_layer_sizes=[n_hidden_nodes, n_hidden_nodes],
                            alpha=alpha, max_iter=5000)
        mlp.fit(X_train, y_train)
        plot_2d_separator(mlp, X_train, fill=True, alpha=.3, ax=ax)
        ax.scatter(X_train[y_train==0, 0], X_train[y_train==0, 1], marker="o",alpha=0.3)
        ax.scatter(X_train[y_train==1, 0], X_train[y_train==1, 1], marker="^",alpha=0.3)        
        ax.set_title(f"n_hidden=[{n_hidden_nodes}, {n_hidden_nodes}] \n alpha={alpha : .4f}")
```

### 实例: 乳腺癌良性恶性肿瘤识别

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)
```

```python
min_training = X_train.min(axis=0)
rng_training = X_train.max(axis=0) - X_train.min(axis=0)

X_train_scaled = (X_train - min_training) / rng_training
X_test_scaled  = (X_test  - min_training) / rng_training
```

```python
import pandas as pd
import seaborn as sns

df = pd.DataFrame(np.c_[y_train, X_train_scaled], columns=["target", *cancer.feature_names])

df =df.melt(id_vars=["target"], var_name='features')

plt.figure(figsize=(12, 4), dpi=120)
sns.violinplot(x="features", y="value", hue="target", data=df, 
               inner="quartile", alpha=0.9, bw=0.1, linewidth=0.8,
               palette=["green", "orange"],  ax=plt.gca());

plt.xticks(range(0, len(cancer.feature_names)), cancer.feature_names, 
           rotation=45, horizontalalignment='right', fontsize=8);
plt.tight_layout()
```

下面会显示警告信息，计算因达到最大迭代次数而提前中止，所得到的结论可能未必是最优的
> `Maximum iterations (200) reached and the optimization hasn't converged yet.`



```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(hidden_layer_sizes=(2,))
model.fit(X_train_scaled, y_train)

from sklearn.metrics import accuracy_score

pred_train = model.predict(X_train_scaled)
pred_test  = model.predict(X_test_scaled)

print(f"Accuracy on training set: { accuracy_score(y_train, pred_train) :.3f}")
print(f"Accuracy on training set: { accuracy_score(y_test, pred_test) :.3f}")
```

调高`max_iter`

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(hidden_layer_sizes=(2,), random_state=0, solver="lbfgs", max_iter=1000)
model.fit(X_train_scaled, y_train)

from sklearn.metrics import accuracy_score

pred_train = model.predict(X_train_scaled)
pred_test  = model.predict(X_test_scaled)

print(f"Accuracy on training set: { accuracy_score(y_train, pred_train) :.3f}")
print(f"Accuracy on training set: { accuracy_score(y_test, pred_test) :.3f}")
```

调整`alhpa`

```python
model = MLPClassifier(hidden_layer_sizes=(100,), alpha=2,  solver="lbfgs", max_iter=1000)
model.fit(X_train_scaled, y_train)

from sklearn.metrics import accuracy_score

pred_train = model.predict(X_train_scaled)
pred_test  = model.predict(X_test_scaled)

print(f"Accuracy on training set: { accuracy_score(y_train, pred_train) :.3f}")
print(f"Accuracy on training set: { accuracy_score(y_test, pred_test) :.3f}")
```
